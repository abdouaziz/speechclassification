```python
import librosa
```

```python
import librosa.display
```

```python
file='a.mp3'
```

```python
signal ,sr =  librosa.load(file , sr=22050)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 


librosa.display.waveplot(signal , sr)
plt.xlabel('Amplitude')
plt.ylabel('Time')
plt.show()
```

```python
#Spectrum
```

```python
fft=  np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0 , sr , len(magnitude))
plt.plot(frequency , magnitude)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.show()
```

```python
#Here We capture the left spectrum

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude= magnitude[:int(len(magnitude)/2)]
plt.plot(left_frequency , left_magnitude)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.show()
```

```python
# SFTF -----> SPECTROGRAM

n_fft= 2048
hop_lenght = 512

stft =  librosa.core.stft(signal , hop_length=hop_lenght , n_fft=n_fft)
spectogram = np.abs(stft)

log_spectogram = librosa.amplitude_to_db(spectogram)
librosa.display.specshow(log_spectogram , sr= sr , hop_length=hop_lenght)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
plt.show()
```

```python
#Here we extract to the MFFCs 

MFFCs = librosa.feature.mfcc(signal , n_fft=n_fft , hop_length= hop_lenght  , n_mfcc=13)
librosa.display.specshow(MFFCs , sr= sr , hop_length=hop_lenght)
plt.xlabel('Time')
plt.ylabel('MFFCs')
plt.colorbar()
plt.show()
```
